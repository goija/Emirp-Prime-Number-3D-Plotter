```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lambertw # Nodig voor de exacte oplossing

# Bereken de exacte oplossing T voor T^T = 49 met de Lambert W-functie
# Dit is voor het markeren van het punt op de grafiek
log_49 = np.log(49)
# lambertw retourneert een complex getal; neem het reÃ«le deel voor T
T_solution = np.real(np.exp(lambertw(log_49)))

# Definieer het bereik voor T voor de plot
# We beginnen T iets boven 0, omdat 0^0 een speciaal geval is (limiet is 1).
# We laten T lopen tot 3.5 om de intersectie goed te kunnen zien zonder dat de y-as te groot wordt.
T_values = np.linspace(0.01, 3.5, 400)
Y_values = T_values**T_values

# Maak de plot aan
plt.figure(figsize=(10, 6))

# Plot de functie T^T
plt.plot(T_values, Y_values, label='$Y = T^T$')

# Plot de horizontale lijn Y = 49
plt.axhline(y=49, color='r', linestyle='--', label='$Y = 49$')

# Markeer het snijpunt (de oplossing)
# T_solution ** T_solution zou heel dicht bij 49 moeten liggen
plt.plot(T_solution, T_solution**T_solution, 'go', markersize=8, label=f'Oplossing ($T \\approx {T_solution:.3f}$)')

# Titels en labels
plt.title('Grafiek van $Y = T^T$ en de oplossing voor $T^T = 49$')
plt.xlabel('$T$')
plt.ylabel('$Y$')
plt.legend()
plt.grid(True)

# Stel de limieten van de assen in voor een duidelijke weergave
plt.xlim(0, 3.5)
plt.ylim(0, 100) # 3.5^3.5 is ongeveer 90.7, dus 100 is een goede y-limiet

# Sla de grafiek op (optioneel, voor als plt.show() niet direct werkt in de omgeving)
# plot_filename = "TT_grafiek.png"
# plt.savefig(plot_filename)

# Toon de plot
plt.show()
```
